c- Trendline : 
You can add a trendline as in the following :

**trendline(str)**- L'une des options suivantes : "ols", "lowess", "rolling", "expanding" ou "ewm". Si 'ols', une ligne de rgression des moindres carrs ordinaires sera trace pour chaque groupe de couleurs/symboles discrets. Si " lowess", une ligne de lissage du nuage de points pondr localement sera trace pour chaque groupe de couleurs/symboles discrets. Si "rolling", une ligne "rolling" (par exemple, moyenne "rolling", mdiane "rolling") sera trace pour chaque groupe de couleurs/symboles discrets. Si "expansion", une ligne d'expansion (par exemple, moyenne d'expansion, somme d'expansion) sera trace pour chaque groupe de couleurs/symboles discrets. Si 'ewm', une ligne de Moment exponentiellement pondr (par exemple, moyenne mobile exponentiellement pondre) sera trace pour chaque groupe de couleurs/symboles discrets. Voir la documentation des fonctions dans plotly.express.trendline_functions pour plus de dtails sur ces fonctions et comment les configurer avec l'argument trendline_options.

```python
fig = px.scatter(df, x = "sepal_width", y = "sepal_length", color = "species", trendline = "ols")
fig.show()
```

e- Marginals : 
Let's go back to the 2D scatter plot. You can add some summarized information by using the marginal_x and marginal_y arguments 

**marginal_x(str)** - L'un des lments suivants : "rug", "box", "violin" ou "histogram". Si cette option est active, un sous-graphe horizontal est dessin au-dessus du graphe principal, visualisant la distribution des x.

**marginal_y(str)** - L'une des options suivantes : "rug", "box","violin " ou "histogram". Si cette option est active, un sous-graphe vertical est trac  droite du graphique principal, afin de visualiser la distribution en y.

```python
fig = px.scatter(df, x = "sepal_width", y = "sepal_length", color = "species", trendline = "ols", marginal_x = "violin", marginal_y = "box")
fig.show()
```

f- Facets : 
You can use facet_row and facet_col to create some facets, each containing the data for a given category. 

If you're using it together with the marginal_x and marginal_y options, the marginals won't be displayed

**facet_row (str ou int ou Series ou array-like)** - Soit le nom d'une colonne dans le data_frame, soit un objet pandas Series ou array_like. Les valeurs de cette colonne ou de cet objet array_like sont utilises pour attribuer des marques aux sous-graphes  facettes dans le sens vertical.

**facet_col (str ou int ou Series ou array-like)** - Soit le nom d'une colonne dans le data_frame, soit un objet pandas Series ou array_like. Les valeurs de cette colonne ou de cet objet array_like sont utilises pour attribuer des marques aux sous-graphes  facettes dans le sens horizontal.

**facet_col_wrap(int)** - Nombre maximal de colonnes  facettes. Enveloppe la variable colonne  cette largeur, de sorte que les facettes de la colonne s'tendent sur plusieurs lignes. Ignor si 0, et forc  0 si facet_row ou un marginal est dfini.

**facet_row_spacing(float between 0 and 1)** - Espacement entre les lignes de facettes, en units de papier. La valeur par dfaut est 0.03 ou 0.0.7 lorsque facet_col_wrap est utilis.

**facet_col_spacing(float between 0 and 1)** - Espacement entre les colonnes de facettes, en units de papier. La valeur par dfaut est de 0.02.

```python
# Create one subgraph for each category and set the display order
fig = px.scatter(df, x = "sepal_width", 
                 y = "sepal_length", 
                 color = "species", 
                 trendline = "ols",
                marginal_x = "violin", 
                 marginal_y = "box",
                 facet_row = "species")
fig.show()
```

```python
# Create one subgraph for each category and set the dislay order
fig = px.scatter(df, x = "sepal_width", 
                 y = "sepal_length", 
                 color = "species", 
                 trendline = "ols", 
                 marginal_x = "violin", 
                 marginal_y = "box", 
                 facet_row = "species", 
                 facet_col = "species_id")
fig.show()
```

    5.With plotly.graph_objects, create a histogram with a range slider such that you can visualize the daily number of events

```python
# Create a figure with plotly
fig = go.Figure()

# Add histogram trace
fig.add_trace(go.Histogram(x=df['Date'], histfunc='count'))

# Update layout to add a range slider
fig.update_layout(
    title='Daily Number of Events',
    xaxis=dict(
        title='Date',
        type='date',
        tickformat='%Y-%m',
        rangeslider=dict(
            visible=True
        ),
    ),
    yaxis=dict(
        title='Number of Events'
    )
)

# Show the plot
fig.show()
```

```python
fig=go.Figure(
    data=go.histogram(df, x=df['Date']),
    layout=go.Layout(title=go.layout.Title(text = "Number of observations per day", x=0.5 ))
)
fig.show()
```
